```
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout,Conv2D,Flatten,Dense,Activation,MaxPooling2D,ZeroPadding2D
from keras.preprocessing.image import ImageDataGenerator
from keras import optimizers
import scipy
import pylab as pl
import matplotlib.cm as cm
%matplotlib inline 
```

```
input_shape=(150,150,3)
img_width=150
img_height=150
nb_train_samples=166
nb_validation_samples=72
batch_size=16
epochs=25

train_data_dir="/content/base/training_data"
nb_validation_data_dir="/content/base/validation_data"
```

```
validation_datagen=ImageDataGenerator(
    rescale=1./255
)
train_datagen=ImageDataGenerator(
    rescale=1./255,
    rotation_range =30,
    width_shift_range=0.3,
    height_shift_range=0.3,
    horizontal_flip=True,
    fill_mode="nearest")

 
```

```
train_generator=train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_width,img_height),
    batch_size=batch_size,
    class_mode="binary",
    shuffle=True)

validation_generator=validation_datagen.flow_from_directory(
    nb_validation_data_dir,
    target_size=(img_width,img_height),
    batch_size=batch_size,
    class_mode="binary",
    shuffle=True)
```

```
model=Sequential()
model.add(Conv2D(32,kernel_size=(3,3),activation='relu',input_shape=input_shape))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(2))

model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())    

```

```
history=model.fit_generator(
    train_generator,
    steps_per_epoch=nb_train_samples // batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=nb_validation_samples // batch_size
)
```
